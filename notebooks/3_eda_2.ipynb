```python
import pandas as pd

# -----------------------------
# 1. Standardize column names
# -----------------------------

# For df1, columns are already in a friendly format.
df1 = df1.rename(
    columns={
        "Уровень 2": "Уровень2",
        "Уровень 3": "Уровень3",
        "Уровень 4": "Уровень4",
        "Уровень 5": "Уровень5",
    }
)

# For df2, rename IT-level columns.
df2 = df2.rename(
    columns={
        "Направление ИТ-услуги. Уровень 2.": "Уровень2",
        "Направление ИТ-услуги. Уровень 3": "Уровень3",
        "Направление ИТ-услуги. Уровень 4.": "Уровень4",
        "Направление ИТ-услуги. Уровень 5.": "Уровень5",
    }
)

# For df3, rename columns (assuming the order corresponds as follows).
df3 = df3.rename(
    columns={
        "2ур": "Уровень2",
        "4ур": "Уровень3",
        "4ур.1": "Уровень4",
        "5ур": "Уровень5",
    }
)

# For df4, remove extra spaces/newlines and rename.
df4 = df4.rename(
    columns={
        "Направление ИТ-услуги.\n Уровень 2. Плашка": "Уровень2",
        "Направление ИТ-услуги.\n Уровень 3. Плашка": "Уровень3",
        "Направление ИТ-услуги. Уровень 4.\nФорма с выпадающими меню на основе уровня 1-3": "Уровень4",
        "Направление ИТ-услуги. Уровень 5.\nФорма с выпадающими меню на основе уровня 1-4": "Уровень5",
    }
)

# Define a common column order (assuming these are all the columns you need)
common_cols = [
    "Номер",
    "Когда создано",
    "Тема",
    "Описание",
    "Услуга",
    "Таблица",
    "Тэг",
    "Назначено на группу",
    "Тип контакта",
    "Уровень2",
    "Уровень3",
    "Уровень4",
    "Уровень5",
]

# Reorder columns (if needed)
df1 = df1[common_cols]
df2 = df2[common_cols]
df3 = df3[common_cols]
df4 = df4[common_cols]


df4 = df4[df4["Уровень2"].notna()]
df3 = df3[df3["Уровень2"].notna()]
df2 = df2[df2["Уровень2"].notna()]
df1 = df1[df1["Уровень2"].notna()]

df = pd.concat([df1, df2, df3, df4])
df = df.drop_duplicates()
```

```python
df.to_csv("../data/raw_1/merged_df.csv", index=False)
```
