## Merge

```python
import pandas as pd

# -----------------------------
# 1. Standardize column names
# -----------------------------

# Для df1 переименовываем столбцы
df1 = df1.rename(
    columns={
        "Уровень 2": "Уровень2",
        "Уровень 3": "Уровень3",
        "Уровень 4": "Уровень4",
        "Уровень 5": "Уровень5",
    }
)

# Для df2 переименовываем IT-level столбцы
df2 = df2.rename(
    columns={
        "Направление ИТ-услуги. Уровень 2.": "Уровень2",
        "Направление ИТ-услуги. Уровень 3": "Уровень3",
        "Направление ИТ-услуги. Уровень 4.": "Уровень4",
        "Направление ИТ-услуги. Уровень 5.": "Уровень5",
    }
)

# Для df3 переименовываем столбцы (согласно порядку)
df3 = df3.rename(
    columns={
        "2ур": "Уровень2",
        "4ур": "Уровень3",
        "4ур.1": "Уровень4",
        "5ур": "Уровень5",
    }
)

# Для df4 убираем лишние символы и переименовываем
df4 = df4.rename(
    columns={
        "Направление ИТ-услуги.\n Уровень 2. Плашка": "Уровень2",
        "Направление ИТ-услуги.\n Уровень 3. Плашка": "Уровень3",
        "Направление ИТ-услуги. Уровень 4.\nФорма с выпадающими меню на основе уровня 1-3": "Уровень4",
        "Направление ИТ-услуги. Уровень 5.\nФорма с выпадающими меню на основе уровня 1-4": "Уровень5",
    }
)

# Общий перечень столбцов, необходимых для объединения
common_cols = [
    "Номер",
    "Когда создано",
    "Тема",
    "Описание",
    "Услуга",
    "Таблица",
    "Тэг",
    "Назначено на группу",
    "Тип контакта",
    "Уровень2",
    "Уровень3",
    "Уровень4",
    "Уровень5",
]

# Оставляем только нужные столбцы
df1 = df1[common_cols]
df2 = df2[common_cols]
df3 = df3[common_cols]
df4 = df4[common_cols]

# Фильтруем строки с наличием данных в столбце "Уровень2"
df1 = df1[df1["Уровень2"].notna()]
df2 = df2[df2["Уровень2"].notna()]
df3 = df3[df3["Уровень2"].notna()]
df4 = df4[df4["Уровень2"].notna()]

# -----------------------------
# 2. Добавление и обработка df5
# -----------------------------

# Переименовываем столбцы для соответствия общему формату
df5 = df5.rename(
    columns={
        "Столбец1": "Уровень3",
        "Столбец2": "Уровень4",
        "Столбец3": "Уровень5",
    }
)

# Создаем недостающие столбцы
df5["Уровень2"] = pd.NA
df5["Когда создано"] = df5["Когда создано"].astype(str)
for col in ["Услуга", "Таблица", "Тэг", "Тип контакта"]:
    df5[col] = pd.NA

# Оставляем только нужные столбцы и фильтруем по наличию данных в "Уровень2"
df5 = df5[common_cols]

# -----------------------------
# 3. Объединение датафреймов
# -----------------------------
df = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)
# df = df.drop_duplicates(subset=["Номер"])

# -----------------------------
# 4. Сборка итоговой информации
# -----------------------------

# Формируем итоговые поля
df["id"] = df["Номер"]
df["class"] = df["Уровень2"]
df["description"] = df["Описание"]
df["subject"] = df["Тема"]

# Логика формирования поля "task"
# Если в "Уровень5" нет данных или стоит "-", то берем значение из "Уровень4",
# а если и там пусто, то из "Уровень3"
df["task"] = df["Уровень5"]
df.loc[df["task"].isna() | (df["task"] == "-"), "task"] = df["Уровень4"]
df.loc[df["task"].isna() | (df["task"] == "-"), "task"] = df["Уровень3"]
df = df[df["task"].notna()]

# Оставляем в итоговом датафрейме только необходимые поля
final_cols = ["id", "description", "subject", "task"]
# df = df[final_cols]
```
