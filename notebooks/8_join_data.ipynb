```python
import pandas as pd

# Пути к вашим четырём файлам
files = [
    '/home/vadim/vadim/request_classifier/data/itsm_task_20250204123503 v2 [Eu9bv0].xlsx',
    '/home/vadim/vadim/request_classifier/data/itsm_task_20250204123503_v2_с_правками.xls',
    '/home/vadim/vadim/request_classifier/data/Разметка данных_ГПСЭД (1).xls',
    '/home/vadim/vadim/request_classifier/data/Разметка_ОМР v2.xlsx'
]

# Список для обработки
processed = []

for path in files:
    df = pd.read_excel(path)
    
    # Определяем имя последней колонки
    last_col = df.columns[-1]
    pre_last_col = df.columns[-2]
    
    # Выбираем нужные столбцы и переименовываем последний
    df_small = df[['Номер', 'Тема', 'Описание', pre_last_col, last_col]].copy()
    df_small = df_small.rename(columns={last_col: 'service_code', pre_last_col: 'service_path'})
    
    # Отбрасываем записи без кода услуги
    df_small = df_small.dropna(subset=['service_code'])
    processed.append(df_small)

# Объединяем воедино
merged_df = pd.concat(processed, ignore_index=True)

# Сохраняем, если нужно
merged_df.to_excel('/home/vadim/vadim/request_classifier/data/merged_services.xlsx', index=False)

# Или просто возвращаем результат
merged_df.head(10)
```
